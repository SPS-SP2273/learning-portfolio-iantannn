# 2 Comparing the two ‘dialects’

```python
#create 100 points from -pi to pi
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

### pyplot Interface

```python
#language of pyplot

plt.plot(x, cos_x, label = "cos x")
plt.plot(x, sin_x, label = "sin x")
plt.legend()
```

### OO Interface

```python
#language of OO

#create a figure called "fig",
#with an axis called "ax", where there is
#1 column (y-axis)and 1 row (x-axis)
fig, ax = plt.subplots(nrows=1, ncols=1)

#plot the axis containing 1 x and y axis
ax.plot(x, cos_x, label="cos x")
ax.plot(x, sin_x, label="sin x")

#plot the legend for each plot made
ax.legend()

plt.show()
```

```python
fig , ax = plt.subplots(nrows=2, ncols=1)
ax[0].plot(x,cos_x, label="cosx")
ax[0].legend()
ax[1].plot(x,sin_x, label="sinx")
ax[1].legend()
plt.show()

#when nrows/ncols becomes more than 1, ax becomes an array
#need to specify which axis you are plotting by indexing it
```

# 4 A complete OO example

```python
#figsize shows the dimensions of the graph being 5 by
#5 inches
#sharex=True means they share the same x-axis,
#important for plots with different x labels
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),)

#$\$ denotes a mathematical LaTeX expression
ax[0].plot(x, cos_x, label=r'$\cos(x)$')

#.fil.between fills area between cos(x) and 0 on y-axis
#alpha=.25 suggests transparency by 0.25
ax[0].fill_between(x, 0, cos_x, alpha=.25)

#the same arguments for element [1]
ax[1].plot(2*x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

#for loop to define the legend, grid and label the y-axis "y"
for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

#only the bottom subplot is labelled
ax[1].set_xlabel('$x$')

#suptitle specifies the title
#fig.tight_layout()
fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
```

```python
#figsize shows the dimensions of the graph being 5 by
#5 inches
#sharex=True means they share the same x-axis
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex="col")

#$\$ denotes a mathematical LaTeX expression
ax[0].plot(x, cos_x, label=r'$\cos(x)$')

#.fil.between fills area between cos(x) and 0 on y-axis
#alpha=.25 suggests transparency by 0.25
ax[0].fill_between(x, 0, cos_x, alpha=.25)

#the same arguments for element [1]
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

#for loop to define the legend, grid and label the y-axis "y"
for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

#only the bottom subplot is labelled
ax[1].set_xlabel('$x$')

#suptitle specifies the title
#fig.tight_layout() minimize the overlap of the axes
fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
```

```python
#figsize shows the dimensions of the graph being 5 by
#5 inches
#sharex=True means they share the same x-axis
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=False)

#r$\$ denotes a mathematical LaTeX expression
ax[0].plot(x, cos_x, label=r'$\cos(x)$')

#.fil.between fills area between cos(x) and 0 on y-axis
#alpha=.25 suggests transparency by 0.25
ax[0].fill_between(x, 0, cos_x, alpha=.25)

#the same arguments for element [1]
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

#for loop to define the legend, grid and label the y-axis "y"
for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

#only the bottom subplot is labelled
ax[1].set_xlabel('$x$')

#suptitle specifies the title
#fig.tight_layout()
fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
```

# 5 Multiple rows and columns

### Using ax

```python
#ax contains 2 rows and 2 columns
#sharex, sharey = "col" means to align at the column
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
# 2x2 NumPy array of axes into indiv variable
top_left, top_right, bottom_left, bottom_right = ax.flatten()

#set a title for each element in the array
top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

#plot a horizontal line, y=1, {0<x<4}
top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

#sets the grid lines for all elements in the array
for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

# 6 Other useful plots

## 6.1 Histograms

### Data

### A quick helper function

```python
def get_plot_data():
    #plot the data into a dictionary
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    #load the data set, into the key "ALL"
    data['All'] = np.loadtxt(filename, skiprows=1)

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    #load the data set into the key "NUS"
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data

```

### The histogram

```python
#"bmh" specific matplotlib style
plt.style.use('default')
#assigned the data of the files into the variable
#"data"
data = get_plot_data()

#bins specifies how many bins to split the data
#plot the histogram using the dataset from All and NUS
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])
plt.xlabel('Mean of Basic Monthly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
```

## 6.2 Scatter plots

```python
plt.style.use("default")

data = get_plot_data()

for label, numbers in data.items():
    x = range(len(numbers))
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
```

## 6.3 Bar charts

```python
#Dictionary containing the number of students and the subject
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
#compile each major(dictionary keys) into a list 
majors = list(student_numbers.keys())

#Compile the frequency (values) into a list
numbers = list(student_numbers.values())

#style of the plot set as default
plt.style.use('default')

#plot a bar chart, x axis: major, y axis: numbers
#(plt.bar(x-axis, y-axis))
plt.bar(majors, numbers)

#labels for x and y axis
plt.xlabel('Majors')
plt.ylabel('Number of Students')
```

## 6.4 Pie charts

```python
#dictionary containing keys (subjects) and values (frequency)
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}

#keys are the majors
majors = list(student_numbers.keys())

#values are the number of students
numbers = list(student_numbers.values())

plt.style.use('default')
plt.pie(numbers, 
        labels=majors,
        
        #autopct formats how the percentages are shown
        #one digit before the decimal and 1 digit after
        autopct='%1.1f%%',   # How to format the percentages
        
        #"-"means to start anticlockwise, and 90 means
        #it starts from the top
        startangle=-90                
        )
plt.title('Percentage of each major')
```
