# 1 A simple plot

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

#plotting the graph
#supported linestyles:
#'-', '--', '-.', ':', 'None', ' ', '', 'solid',
#'dashed', 'dashdot', 'dotted'
#linewidth default 1 (?) if not specified

plt.plot(x, y_1, label = "y=2x", color = "salmon", 
         linestyle = "solid")

plt.xlabel("x-values")
#y-axis label
plt.ylabel("y-values")
#title of the graph
plt.title("X vs Y")
#opacity of the axis 0-1
plt.grid(alpha=0.3)
#'best', 'upper right', 'upper left', 'lower left', 
#'lower right', 'right', 'center left', 'center right'
#'lower center', 'upper center', 'center'
plt.legend(loc = "upper left")
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

#plotting the graph
plt.plot(x, y_1, label = "y=2x", color = "purple", 
         linestyle = "dotted",linewidth = 2)
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("X vs Y")
plt.grid(alpha=0.1)
plt.legend(loc = "center")
```

## 1.2 Adding another plot

```python
plt.plot(x, y_2, 
        label = "Y2 values", c = "blue")
plt.plot(x,y_1,
        label = "Y1 values", c = "magenta")
plt.legend(loc = "upper left")
plt.show()
```

## 1.3 Yet another plot but with error bars

```python
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.errorbar(x, y_3,
            xerr=x_error, yerr=y_error,
            label = "Y3 with errors",
            c = "cyan")
plt.plot(x, y_1, label="Y1 values")
plt.plot(x, y_2, label="Y2 values")

plt.grid(alpha=.5)
plt.legend(loc="upper left")
```

# 2 Better with NumPy

```python
import numpy as np
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, ls="solid", c = "blue", label="Y1")
plt.plot(np_x, np_y_2, ls="solid",c = "red", label="Y2")
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 2.1 Adding mathematical functions

```python
x = []
for i in range(7):
    x.append(i)
    
np_x=np.array(x)
    


x2 = np_x**2
sin_x = np.sin(np_x)

plt.plot(x, x2, label = "x^2", linestyle = "solid")
plt.plot(x,sin_x, label = "sin(x)", linestyle = "solid")
plt.legend()
plt.grid(alpha=0.25)
```

## We need another axis!

```python
x = []
for i in range(7):
    x.append(i)


np_x=np.array(x)

x2 = np_x**2
sin_x = np.sin(np_x)

plt.plot(np_x, x2, label = "x^2", linestyle = "solid", c= "red")
plt.legend(loc="upper right")

#has to be placed between the two plot functions as the
#code is read sequentially

#to indicate a second plot and axis for the other graph
plt.twinx()

plt.plot(np_x,sin_x, label = "sin(x)", linestyle = "solid", c = "blue")
plt.legend(loc="upper left")
plt.grid(alpha=0.25)
plt.show()
```

```python
#generate 100 points into a numpy array from 0 to 6
np_x = np.linspace(0,6,100)


x2 = np_x**2
sin_x = np.sin(np_x)

plt.plot(np_x, x2, label = "x^2", linestyle = "solid", c= "red")
plt.legend(loc="upper right")

#has to be placed between the two plot functions as the
#code is read sequentially

#to indicate a second plot and axis for the other graph
plt.twinx()

plt.plot(np_x,sin_x, label = "sin(x)", linestyle = "solid", c = "blue")
plt.legend(loc="upper left")
plt.grid(alpha=0.25)
plt.show()
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
#loading the file from the computer
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)

#select all rows of the first column "0"
date = data[:, 0]

#select all rows of the last column "-1"
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)

#sets the range of the y axis to be from -8 to 8
plt.ylim([-8, 8])

plt.show()
```

```python
#loading the file from the computer
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)

#select all rows of the first column "0"
date = data[:, 0]

#select all rows of the last column "-1"
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)

#sets the range of the y axis to be from -8 to 8
plt.ylim([-8, 8])

#plotting a horizontal line (hline)
#using the "date" array where the starting x-coord
#is element 0 in the date column and the last element
#being the ending coordinate.
plt.hlines(0, date[0], date[-1], linestyle='--', 
           colors='grey')

plt.show()
```

## 4.2 Styles
