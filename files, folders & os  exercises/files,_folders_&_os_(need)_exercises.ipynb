```python

path = "os-collaboration-exercise-data"


#list with every file path being an element
files = glob.glob(os.path.join(path, "**", "*.txt"), recursive=True)

def rename_my_file(old_file_name):
    length = old_file_name.split("-")
    if len(length) != 3:
        month = old_file_name.split("-")[0]
        rest_of_old_file_name = old_file_name.split("-")[1]
        country = rest_of_old_file_name.split("_")[-1]
        date = rest_of_old_file_name.split("_")[0]
        return f"{dates[month]}-{date}-{country}"

#for loop to run through every path in the list "files"
for file in files:
    #split the file path to just obtain the file name in the list
    filename = file.split(os.path.sep)[-1]
    new_filename = rename_my_file(filename)
    #define the path that the new file will be
    new_path = os.path.join(os.path.dirname(file), new_filename)
    #os.rename(current path, new path )
    os.rename(file, new_path)

    
    
    
```

## Task 7

```python
path = "os-collaboration-exercise-data"
list_of_countries= ['Cairo','Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for country in list_of_countries:
    #create path i.e os-collaboration-exercise-data/Dhaka
    path_to_folder = os.path.join(path,country)
    
    #filenames are lowercase
    filename_country = country.lower()
    
    #globals()[] is a function that helps convert the string to a variable instead
    #each string in the list_of_countries now becomes a variable containing a list 
    # which contains the respective files
    #Dhaka = os-collaboration-exercise-data/07-26_Dhaka.txt
    globals()[country]=glob.glob(os.path.join(path, "**",f"*{filename_country}.txt"), recursive = True)
    for file in globals()[country]:
        filename = file.split(os.path.sep)[-1]
        if os.path.exists(os.path.join(path_to_folder, filename)):
            print(f"{file} is already transferred, skipping" )
        else:
            shutil.copy(file, path_to_folder)
            os.remove(file)
            print(f"{file} is transferred")
```
