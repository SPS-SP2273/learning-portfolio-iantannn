## Task 4

```python
import os
import glob
import shutil
```

```python

path = "os-collaboration-exercise-data"


#list with every file path being an element
files = glob.glob(f"{path}/**/*.txt", recursive=True)


#for loop to run through every path in the list "files"
for file in files:
    #split the file path to just obtain the file name in the list
    filename = file.split(os.path.sep)[-1]
    
    #split the individual name to get the month
    split_filename = filename.split("-")
    #split_filename[0] = month
    #split_filename[1] = the remaining half
    #the new list consists ["jan", "14_tokyo"] for example
    
    #define the new filename
    new_filename = f"{dates[split_filename[0]]}-{split_filename[1]}"
    
    #define the path that the new file will be
    new_path = os.path.join(os.path.dirname(file), new_filename)
    
    
    #os.rename(current path, new path )
    os.rename(file, new_path)
    
```
